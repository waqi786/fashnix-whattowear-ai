## Project Overview

Fashnix is an end‑to‑end deep learning system that classifies fashion product images into **multiple attributes simultaneously**:

- **Master Category** (4 robust classes – Accessories, Apparel, Footwear, Personal Care)  
- **Base Colour** (5 classes – **Black, White, Blue, Red, Other**)  
- **Occasion / Usage** (7 classes – Casual, Formal, Sports, Party, etc.)  
- **Gender** (5 classes – Men, Women, Unisex, Boys, Girls)

The model uses a **MobileNetV2** backbone pre‑trained on ImageNet, followed by four custom classification heads.  
The **colour head** receives high‑resolution features from an early MobileNetV2 block (14×14×96) together with a dedicated convolutional branch, ensuring the best possible colour recognition from the available data.

> **Trustworthy Colour Strategy**  
> The original dataset has heavily noisy colour labels (single colour per image, mis‑labellings, multiple colours in one picture). Instead of forcing the model to predict all colours and risk frequent, confident mistakes, we group colours into four unambiguous, frequently occurring classes (**Black, White, Blue, Red**) and one fallback class (**Other**).  
> When the model predicts **“Other”**, the app will ask the user to manually select the colour.  
> This honest design **never confidently shows a wrong colour**, building user trust and long‑term engagement. Future user corrections can be stored and used to shrink the “Other” category progressively.

After rigorous training and fine‑tuning, the model is converted to **TensorFlow Lite** with float16 quantisation for fast, on‑device inference on Android.

```python
# =============================================================================
# CELL 10: Clean the dataset
#          - Keep only samples whose image file exists
#          - Remove extremely rare categories (Home, Sporting Goods, Free Items)
#          - Map the 46 original colours into 5 trustworthy classes:
#            Black, White, Blue, Red, Other
# =============================================================================

image_dir = "/kaggle/input/datasets/paramaggarwal/fashion-product-images-small/images"

# ---------- 1. Image existence ----------
def image_exists(row):
    img_path = os.path.join(image_dir, f"{int(row['id'])}.jpg")
    return os.path.isfile(img_path)

print("Checking image existence …")
mask = df.apply(image_exists, axis=1)
df = df[mask].copy()
print(f"Dropped {(~mask).sum()} rows due to missing images → {df.shape[0]} rows remain")

# ---------- 2. Drop extremely rare categories ----------
drop_categories = ['Home', 'Sporting Goods', 'Free Items']
initial_rows = df.shape[0]
df = df[~df['masterCategory'].isin(drop_categories)].copy()
print(f"Dropped {initial_rows - df.shape[0]} rows from {drop_categories}")
print(f"Remaining samples: {df.shape[0]}")

# ---------- 3. Map original 46 colours → 5 trustworthy classes ----------
# Define mapping: Black, White, Blue, Red = keep; everything else = Other
colour_map = {
    'Black': 'Black', 'Charcoal': 'Black', 'Jet Black': 'Black',
    'White': 'White', 'Cream': 'White', 'Off White': 'White', 'Taupe': 'White',
    'Sepia': 'White', 'Khaki': 'White', 'Tan': 'White', 'Nude': 'White',
    'Skin': 'White', 'Beige': 'White', 'Beige & Black': 'Black',
    'Blue': 'Blue', 'Navy Blue': 'Blue', 'Turquoise Blue': 'Blue',
    'Teal': 'Blue', 'Sea Green': 'Blue',
    'Red': 'Red', 'Maroon': 'Red', 'Burgundy': 'Red', 'Rust': 'Red',
    'Coral': 'Red', 'Peach': 'Red', 'Rose': 'Red', 'Magenta': 'Red',
    'Mauve': 'Red', 'Lavender': 'Red', 'Purple': 'Red',
    # Everything else → Other
    'Pink': 'Other', 'Rose Gold': 'Other',
    'Green': 'Other', 'Olive': 'Other', 'Lime Green': 'Other',
    'Fluorescent Green': 'Other',
    'Grey': 'Other', 'Silver': 'Other', 'Tin': 'Other', 'Grey Melange': 'Other',
    'Brown': 'Other', 'Coffee Brown': 'Other', 'Bronze': 'Other',
    'Copper': 'Other', 'Mushroom Brown': 'Other',
    'Orange': 'Other', 'Yellow': 'Other', 'Gold': 'Other', 'Mustard': 'Other'
}

df['baseColour'] = df['baseColour'].map(colour_map)
df = df.dropna(subset=['baseColour']).copy()

print(f"After colour mapping: {df.shape[0]} rows, {df['baseColour'].nunique()} unique colours")
print("Colour distribution:\n", df['baseColour'].value_counts())

print("\nCategory distribution after cleaning:")
print(df['masterCategory'].value_counts())
```

```python
# =============================================================================
# CELL 11: Convert text labels to integers using LabelEncoder.
#          We fit one encoder per task and store them for later inverse mapping.
#          The encoded columns are added to the DataFrame for easy splitting.
# =============================================================================

from sklearn.preprocessing import LabelEncoder

# Define the four tasks and their corresponding DataFrame columns
task_columns = {
    'category': 'masterCategory',
    'colour':   'baseColour',       # now 5 classes: Black, White, Blue, Red, Other
    'occasion': 'usage',
    'gender':   'gender'
}

label_encoders = {}      # store fitted encoders
encoded_labels = {}      # store integer arrays

for task_name, col in task_columns.items():
    df[col] = df[col].astype(str)   # ensure all values are strings
    le = LabelEncoder()
    encoded = le.fit_transform(df[col])
    label_encoders[task_name] = le
    encoded_labels[task_name] = encoded
    print(f"{task_name} ({col}): {len(le.classes_)} classes — {list(le.classes_)}")

# Add encoded columns to DataFrame
df['category_label'] = encoded_labels['category']
df['colour_label']   = encoded_labels['colour']
df['occasion_label'] = encoded_labels['occasion']
df['gender_label']   = encoded_labels['gender']

print("\nSample encoded rows:")
df[['masterCategory', 'category_label',
    'baseColour', 'colour_label',
    'usage', 'occasion_label',
    'gender', 'gender_label']].head()
```

```python
# =============================================================================
# CELL 12: Split data into training (70%), validation (15%), test (15%).
#          Stratification preserves the proportion of each category in all splits.
#          All 4 classes have many samples so stratification works fine.
# =============================================================================

from sklearn.model_selection import train_test_split

# First split: separate 15% as the test set
train_val_df, test_df = train_test_split(
    df,
    test_size=0.15,
    stratify=df['category_label'],
    random_state=42
)

# Second split: from the remaining 85%, separate ~17.65% → 15% of total for validation
val_ratio = 0.1765
train_df, val_df = train_test_split(
    train_val_df,
    test_size=val_ratio,
    stratify=train_val_df['category_label'],
    random_state=42
)

# Print the resulting sizes
print(f"Training set:   {len(train_df)} samples ({len(train_df)/len(df)*100:.1f}%)")
print(f"Validation set: {len(val_df)} samples ({len(val_df)/len(df)*100:.1f}%)")
print(f"Test set:       {len(test_df)} samples ({len(test_df)/len(df)*100:.1f}%)")

# Confirm that category proportions are consistent across splits
for name, subset in [("Train", train_df), ("Val", val_df), ("Test", test_df)]:
    dist = subset['masterCategory'].value_counts(normalize=True).map(lambda x: f"{x:.2%}")
    print(f"\n{name} — category proportions:")
    print(dist)
```

```python
# =============================================================================
# CELL 13: Build TensorFlow Dataset objects for train/val/test.
#          Images are loaded, resized to 224×224, normalized to [0,1].
#          Label dict keys are renamed to match model output layer names.
#          Data augmentation will be added as Keras layers inside the model.
# =============================================================================

IMG_SIZE   = (224, 224)
BATCH_SIZE = 64
image_dir  = "/kaggle/input/datasets/paramaggarwal/fashion-product-images-small/images"

# ---------- Helper: convert DataFrame rows to paths and label dicts ----------
def df_to_paths_and_labels(subset_df):
    """Return list of image paths and a dict of integer label arrays."""
    paths = [os.path.join(image_dir, f"{int(row['id'])}.jpg")
             for _, row in subset_df.iterrows()]
    labels = {
        'category': subset_df['category_label'].values,
        'colour':   subset_df['colour_label'].values,
        'occasion': subset_df['occasion_label'].values,
        'gender':   subset_df['gender_label'].values
    }
    return paths, labels

# ---------- Image loading & preprocessing function ----------
def load_and_preprocess(path, labels):
    """Read, decode, resize, and normalise one image."""
    img = tf.io.read_file(path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, IMG_SIZE)
    img = tf.cast(img, tf.float32) / 255.0
    return img, labels

# ---------- Create path lists and label dicts for each split ----------
train_paths, train_labels = df_to_paths_and_labels(train_df)
val_paths,   val_labels   = df_to_paths_and_labels(val_df)
test_paths,  test_labels  = df_to_paths_and_labels(test_df)

# ---------- Build tf.data.Dataset objects ----------
# Training set: shuffle, batch, prefetch
train_ds = tf.data.Dataset.from_tensor_slices((train_paths, train_labels))
train_ds = train_ds.map(load_and_preprocess, num_parallel_calls=tf.data.AUTOTUNE)
train_ds = train_ds.shuffle(1000)
train_ds = train_ds.batch(BATCH_SIZE)
train_ds = train_ds.prefetch(tf.data.AUTOTUNE)

# Validation set
val_ds = tf.data.Dataset.from_tensor_slices((val_paths, val_labels))
val_ds = val_ds.map(load_and_preprocess, num_parallel_calls=tf.data.AUTOTUNE)
val_ds = val_ds.batch(BATCH_SIZE)
val_ds = val_ds.prefetch(tf.data.AUTOTUNE)

# Test set
test_ds = tf.data.Dataset.from_tensor_slices((test_paths, test_labels))
test_ds = test_ds.map(load_and_preprocess, num_parallel_calls=tf.data.AUTOTUNE)
test_ds = test_ds.batch(BATCH_SIZE)
test_ds = test_ds.prefetch(tf.data.AUTOTUNE)

# ---------- Align label dict keys with model output layer names ----------
def rename_labels(images, labels):
    new_labels = {
        'category_output': labels['category'],
        'colour_output':   labels['colour'],
        'occasion_output': labels['occasion'],
        'gender_output':   labels['gender']
    }
    return images, new_labels

train_ds = train_ds.map(rename_labels)
val_ds   = val_ds.map(rename_labels)
test_ds  = test_ds.map(rename_labels)

print("Datasets created and label keys renamed.")
print(f"  Train batches: {len(train_ds)}")
print(f"  Val batches:   {len(val_ds)}")
print(f"  Test batches:  {len(test_ds)}")
```

```python
# =============================================================================
# CELL 14: Apply random augmentations to a sample batch and display 8 pairs
#          (original vs augmented). Brand theme orange/blue, thick black borders.
# =============================================================================

# --- Augmentation function (same as what the model will use) ---
def augment_image(image):
    """Apply the same augmentations used during training."""
    image = tf.image.random_flip_left_right(image)
    image = tf.keras.layers.RandomRotation(0.2)(image)
    image = tf.keras.layers.RandomZoom(0.2)(image)
    image = tf.image.random_brightness(image, max_delta=0.2)
    image = tf.clip_by_value(image, 0.0, 1.0)
    return image

# Take one batch and select 8 images
for images, labels in train_ds.take(1):
    sample_images = images[:8]
    break

augmented_images = [augment_image(img) for img in sample_images]

# --- Plot 4 rows × 4 columns (original | augmented) ---
fig, axes = plt.subplots(4, 4, figsize=(16, 16))
axes = axes.flatten()

for i in range(8):
    # Original (left column)
    ax_orig = axes[2*i]
    ax_orig.imshow(sample_images[i])
    ax_orig.set_title('Original', fontsize=10, color='#1E3A5F', fontweight='bold')
    ax_orig.axis('off')
    for spine in ax_orig.spines.values():
        spine.set_edgecolor('black')
        spine.set_linewidth(3)

    # Augmented (right column)
    ax_aug = axes[2*i + 1]
    ax_aug.imshow(augmented_images[i])
    ax_aug.set_title('Augmented', fontsize=10, color='#CC5500', fontweight='bold')
    ax_aug.axis('off')
    for spine in ax_aug.spines.values():
        spine.set_edgecolor('black')
        spine.set_linewidth(3)

plt.tight_layout(pad=2.0)
plt.savefig('augmented_samples.jpg', dpi=300, bbox_inches='tight',
            facecolor='white', edgecolor='black')
plt.show()
```

        4. Model Architecture
    </b>
    <span style="font-size: 16px; display: block; font-weight: 500; margin-top: 10px; 
    color: black;">
        We build a Multi-Task CNN using MobileNetV2 to predict category and other details, while using earlier layers to better capture colour information.
This design helps the model understand both clothing type and colour clearly, giving more reliable results.
    </span>

</div>

```python
# =============================================================================
# CELL 16: Multi‑Task CNN with early high‑resolution colour features
#          Colour head uses early MobileNetV2 features (14×14×96) and a
#          dedicated conv branch to preserve spatial colour detail.
#          Output: 5 colour classes (Black, White, Blue, Red, Other)
# =============================================================================

IMG_SHAPE = (224, 224, 3)

# ---------- Data augmentation (as Keras layers) ----------
data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.2),
    layers.RandomZoom(0.2),
    layers.RandomBrightness(0.2),
], name='data_augmentation')

# ---------- Base model (MobileNetV2, frozen for Phase 1) ----------
base_model = MobileNetV2(
    weights='imagenet',
    include_top=False,
    input_shape=IMG_SHAPE
)
base_model.trainable = False

# ---------- Input ----------
inputs = keras.Input(shape=IMG_SHAPE)
aug = data_augmentation(inputs)

# ----- Deep features (semantic) -----
base_features = base_model(aug, training=False)          # (None, 7, 7, 1280)
x_deep = layers.GlobalAveragePooling2D()(base_features)  # (None, 1280)

# ----- Early high‑resolution colour branch -----
# Extract early feature map from block_13_expand_relu (14×14×96)
early_layer_name = 'block_13_expand_relu'
early_feature_model = keras.Model(
    inputs=base_model.input,
    outputs=base_model.get_layer(early_layer_name).output
)
early_features = early_feature_model(aug)               # (None, 14, 14, 96)

# Small conv branch on early features
colour_branch = layers.Conv2D(64, 3, padding='same', activation='relu')(early_features)
colour_branch = layers.MaxPooling2D(2)(colour_branch)          # (None, 7, 7, 64)
colour_branch = layers.Conv2D(128, 3, padding='same', activation='relu')(colour_branch)
colour_branch = layers.GlobalAveragePooling2D()(colour_branch) # (None, 128)
colour_branch = layers.Dense(128, activation='relu')(colour_branch)

# ----- Colour head combines deep context + high‑res colour features -----
colour_combined = layers.Concatenate()([x_deep, colour_branch])  # (None, 1280+128)
col = layers.Dense(256, activation='relu')(colour_combined)
col = layers.BatchNormalization()(col)
col = layers.Dropout(0.3)(col)
colour_output = layers.Dense(len(label_encoders['colour'].classes_),
                             activation='softmax', name='colour_output')(col)

# ----- Category head (deep features only) -----
cat = layers.Dense(256, activation='relu')(x_deep)
cat = layers.BatchNormalization()(cat)
cat = layers.Dropout(0.3)(cat)
category_output = layers.Dense(len(label_encoders['category'].classes_),
                               activation='softmax', name='category_output')(cat)

# ----- Occasion head -----
occ = layers.Dense(128, activation='relu')(x_deep)
occ = layers.Dropout(0.3)(occ)
occasion_output = layers.Dense(len(label_encoders['occasion'].classes_),
                               activation='softmax', name='occasion_output')(occ)

# ----- Gender head -----
gen = layers.Dense(128, activation='relu')(x_deep)
gen = layers.Dropout(0.3)(gen)
gender_output = layers.Dense(len(label_encoders['gender'].classes_),
                             activation='softmax', name='gender_output')(gen)

# ----- Assemble model -----
model = keras.Model(
    inputs=inputs,
    outputs=[category_output, colour_output, occasion_output, gender_output],
    name='Fashnix_TrustworthyColour'
)

print("Model built successfully!")
print(f"  Category head: {len(label_encoders['category'].classes_)} classes")
print(f"  Colour head:   {len(label_encoders['colour'].classes_)} classes")
print(f"  Occasion head: {len(label_encoders['occasion'].classes_)} classes")
print(f"  Gender head:   {len(label_encoders['gender'].classes_)} classes")
```

    <span style="font-size: 16px; display: block; font-weight: 500; margin-top: 10px; 
    color: black;">
        The model is trained in three steps: first training the top layers, then fine-tuning deeper layers, and finally improving colour accuracy.
We use standard techniques like Adam optimizer, learning rate control, and early stopping to get better performance.
    </span>

```python
# =============================================================================
# CELL 19: Compile the multi‑task model.
#          - Loss: weighted sparse_categorical_crossentropy for category,
#                  standard sparse_categorical_crossentropy for others.
#          - Class weights are computed on the training set and capped at 3.0
#            to prevent Personal Care from dominating while still giving it attention.
#          - Optimizer: Adam (lr=0.001)
#          - Metrics: accuracy per task
# =============================================================================

from sklearn.utils.class_weight import compute_class_weight

# --- Compute capped category class weights from the training set ---
y_train_cat = train_df['category_label'].values

# Raw balanced weights: n_samples / (n_classes * bincount)
raw_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train_cat),
    y=y_train_cat
)

# Cap each weight at 3.0 for stable training
capped_weights = np.minimum(raw_weights, 3.0)

# Create a dictionary: class_index → capped weight
class_weight_dict = {i: w for i, w in enumerate(capped_weights)}

# Print the effective weights
print("Category class weights (capped at 3.0):")
for cls_idx, weight in sorted(class_weight_dict.items()):
    class_name = label_encoders['category'].inverse_transform([cls_idx])[0]
    print(f"  {class_name:20s} (class {cls_idx}): raw={raw_weights[cls_idx]:.2f} → capped={weight:.4f}")

# Convert to a constant tensor for the custom loss
class_weight_tensor = tf.constant(
    [class_weight_dict[i] for i in range(len(class_weight_dict))],
    dtype=tf.float32
)

# --- Weighted loss function ---
def weighted_categorical_crossentropy(y_true, y_pred):
    """
    Sparse categorical crossentropy with per‑class sample weights.
    y_true: integer labels [batch]
    y_pred: softmax probabilities [batch, num_classes]
    """
    # Gather the weight for each sample according to its true class
    sample_weights = tf.gather(class_weight_tensor, tf.cast(y_true, tf.int32))
    # Standard sparse categorical crossentropy (no reduction)
    scce = tf.keras.losses.sparse_categorical_crossentropy(
        y_true, y_pred, from_logits=False
    )
    # Weighted mean over the batch
    return tf.reduce_mean(scce * sample_weights)

# --- Compile the model ---
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss={
        'category_output': weighted_categorical_crossentropy,
        'colour_output':   'sparse_categorical_crossentropy',
        'occasion_output': 'sparse_categorical_crossentropy',
        'gender_output':   'sparse_categorical_crossentropy'
    },
    loss_weights={
        'category_output': 1.0,
        'colour_output':   1.0,
        'occasion_output': 1.0,
        'gender_output':   1.0
    },
    metrics={
        'category_output': 'accuracy',
        'colour_output':   'accuracy',
        'occasion_output': 'accuracy',
        'gender_output':   'accuracy'
    }
)

print("\nModel compiled with capped weighted loss (max weight = 3.0).")
```

### Class Weights for Category

The category class weights are capped at **3.0** and integrated into the custom `weighted_categorical_crossentropy` loss.  
This prevents the rare “Personal Care” class from receiving its raw weight of 5.08, which would have caused it to be over‑predicted.  
Now all four classes receive balanced attention without any single class dominating the loss.

The compilation in CELL 19 already uses these weights — no separate cell is needed.

```python
# =============================================================================
# CELL 21: Phase 1 Training — Frozen MobileNetV2 base, only heads train.
#          EarlyStopping (patience=7), ReduceLROnPlateau (factor 0.5, patience 3),
#          best model checkpoint based on val_category_output_accuracy.
#          Plots use orange/blue theme, thick black borders, saved as JPG.
# =============================================================================

# --- Callbacks ---
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=7,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=3,
    min_lr=1e-7,
    verbose=1
)

checkpoint = ModelCheckpoint(
    'best_phase1_model.keras',
    monitor='val_category_output_accuracy',
    mode='max',
    save_best_only=True,
    verbose=1
)

# --- Train Phase 1 ---
print("Starting Phase 1: training heads with frozen base (15 epochs)...\n")
history1 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=15,
    callbacks=[early_stop, reduce_lr, checkpoint],
    verbose=1
)

# --- Plot loss and accuracy ---
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Combined loss
axes[0].plot(history1.history['loss'], color='#FF8C00', linewidth=2.5, label='Train Loss')
axes[0].plot(history1.history['val_loss'], color='#1E90FF', linewidth=2.5, label='Val Loss')
axes[0].set_title('Phase 1 — Combined Loss', fontsize=14, fontweight='bold', color='#1E3A5F')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].legend()
axes[0].grid(True, alpha=0.3)
for spine in axes[0].spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(2)

# Category accuracy
axes[1].plot(history1.history['category_output_accuracy'], color='#FF8C00', linewidth=2.5, label='Train Acc')
axes[1].plot(history1.history['val_category_output_accuracy'], color='#1E90FF', linewidth=2.5, label='Val Acc')
axes[1].set_title('Phase 1 — Category Accuracy', fontsize=14, fontweight='bold', color='#1E3A5F')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Accuracy')
axes[1].legend()
axes[1].grid(True, alpha=0.3)
for spine in axes[1].spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(2)

plt.tight_layout()
plt.savefig('phase1_training_curves.jpg', dpi=300, bbox_inches='tight',
            facecolor='white', edgecolor='black')
plt.show()
```

```python
# =============================================================================
# CELL 22: Phase 2 Fine‑tuning
#          - Load best Phase 1 weights (val_category_accuracy ≈ 92.4%)
#          - Unfreeze the LAST 40 layers of MobileNetV2 base
#          - Re‑compile with learning rate 1e‑4
#          - Train up to 20 epochs
#          - Plot loss & category accuracy (orange/blue theme)
# =============================================================================

# 1. Load best Phase 1 weights
model.load_weights('best_phase1_model.keras')
print("Loaded best Phase 1 weights (val_category_accuracy = 0.9240).\n")

# 2. Unfreeze last 40 layers of the base model
base_model.trainable = True
total_layers = len(base_model.layers)
fine_tune_from = total_layers - 40   # 154 - 40 = 114

# Freeze the first 114 layers, keep the last 40 trainable
for layer in base_model.layers[:fine_tune_from]:
    layer.trainable = False
for layer in base_model.layers[fine_tune_from:]:
    layer.trainable = True

print(f"Base model total layers: {total_layers}")
print(f"Frozen layers: 0 to {fine_tune_from-1} (first {fine_tune_from} layers)")
print(f"Trainable layers: {fine_tune_from} to {total_layers-1} (last 40 layers)")

trainable_now = sum([tf.keras.backend.count_params(w) for w in model.trainable_weights])
print(f"Trainable parameters after unfreezing: {trainable_now:,}\n")

# 3. Recompile with a smaller learning rate
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-4),
    loss={
        'category_output': weighted_categorical_crossentropy,
        'colour_output':   'sparse_categorical_crossentropy',
        'occasion_output': 'sparse_categorical_crossentropy',
        'gender_output':   'sparse_categorical_crossentropy'
    },
    loss_weights={
        'category_output': 1.0,
        'colour_output':   1.0,
        'occasion_output': 1.0,
        'gender_output':   1.0
    },
    metrics={
        'category_output': 'accuracy',
        'colour_output':   'accuracy',
        'occasion_output': 'accuracy',
        'gender_output':   'accuracy'
    }
)

# 4. Callbacks
early_stop2 = EarlyStopping('val_loss', patience=7, restore_best_weights=True, verbose=1)
reduce_lr2 = ReduceLROnPlateau('val_loss', factor=0.5, patience=3, min_lr=1e-7, verbose=1)
checkpoint2 = ModelCheckpoint('best_phase2_model.keras', monitor='val_category_output_accuracy',
                              mode='max', save_best_only=True, verbose=1)

# 5. Train
print("Starting Phase 2 fine‑tuning...\n")
history2 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=20,
    callbacks=[early_stop2, reduce_lr2, checkpoint2],
    verbose=1
)

# 6. Plot curves
fig, axes = plt.subplots(1, 2, figsize=(18, 6))
axes[0].plot(history2.history['loss'], color='#FF8C00', linewidth=2.5, label='Train Loss')
axes[0].plot(history2.history['val_loss'], color='#1E90FF', linewidth=2.5, label='Val Loss')
axes[0].set_title('Phase 2 — Combined Loss', fontsize=14, fontweight='bold', color='#1E3A5F')
axes[0].set_xlabel('Epoch'); axes[0].set_ylabel('Loss'); axes[0].legend(); axes[0].grid(alpha=0.3)
for sp in axes[0].spines.values(): sp.set_edgecolor('black'); sp.set_linewidth(2)

axes[1].plot(history2.history['category_output_accuracy'], color='#FF8C00', linewidth=2.5, label='Train Acc')
axes[1].plot(history2.history['val_category_output_accuracy'], color='#1E90FF', linewidth=2.5, label='Val Acc')
axes[1].set_title('Phase 2 — Category Accuracy', fontsize=14, fontweight='bold', color='#1E3A5F')
axes[1].set_xlabel('Epoch'); axes[1].set_ylabel('Accuracy'); axes[1].legend(); axes[1].grid(alpha=0.3)
for sp in axes[1].spines.values(): sp.set_edgecolor('black'); sp.set_linewidth(2)

plt.tight_layout()
plt.savefig('phase2_training_curves.jpg', dpi=300, bbox_inches='tight', facecolor='white', edgecolor='black')
plt.show()
```
